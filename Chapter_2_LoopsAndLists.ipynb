```python
# Task 1: Convert Degrees to Fahrenheit from -20 to 40
print('--------------')
C = -10
dC = 5
while C <= 40:
    F = (9.0/5)*C + 32
    print(C, F)
    C = C + dC
print('--------------')
```

```python
# Task 2: Approximation of sin(x) using a loop
import math

x = 1.2
n = 25
k = 1
s = x
sign = 1.0

while k < n:
    sign = - sign
    k = k + 2
    term = sign*x**k/math.factorial(k)
    s += term

print(f"sin({x}) = {s} (approximation with {n} terms)")
```
